```python

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}
```

```python
def dfs_recursive(graph, current, goal, visited=None, path=None):
    if visited is None:
        visited = set()

    if path is None:
        path = []

    visited.add(current)
    path.append(current)


    if current == goal:
        return path


    for neighbor in graph[current]:
        if neighbor not in visited:
            result = dfs_recursive(graph, neighbor, goal, visited, path.copy())
            if result:
                return result

    return None
```

```python

start = 'A'
goal = 'F'

result = dfs_recursive(graph, start, goal)

if result:
    print("Path found:", " -> ".join(result))
else:
    print("No path found")
```
